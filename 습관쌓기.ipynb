```python
import pandas as pd
import numpy as np
```

### 기존 데이터프레임 가져오기(.xlsx)

```python
df = pd.read_excel("결과.xlsx",encoding = 'utf-8')
idx_rename = df["Unnamed: 0"]
del df["Unnamed: 0"]
df = df.rename(index = idx_rename)
df.head()
```

### 새로운 날짜 추가

```python
data = open('습관.txt','rt',encoding = 'UTF8')
line = data.readline()
Today = []
count = 0
for x in range(len(line)):
    if(line[x] == '/'):
        break
    elif(line[x] == ']'):
        count += 1
    elif(count == 2):
        Today.append(line[x])
Today = ''.join(Today)
df[Today] = np.nan
df.head()
```

### data 한줄씩 읽고 기록

```python
data = open('습관.txt','rt',encoding = 'UTF8')
lines = data.readlines()
days = df.columns

for x in range(len(lines)):
    name = []
    date = []
    count = 0 #카톡내용중 불필요한 내용 삭제를 위함
    #날짜 저장
    for y in range(len(lines[x])):
        if(lines[x][y] == ']'):
            count += 1
        elif lines[x][y] == '/':
            break
        elif(count == 2):
            date.append(lines[x][y])
    date =''.join(date)
    if not date in days:
        continue
    #이름. 성공.실패여부
    fail = 0
    slash = 0
    for y in range(len(lines[x])):
        if(lines[x][y] == '/'):
            slash += 1
        #이름 저장
        elif slash == 1:
            name.append(lines[x][y])
        #실패횟수
        elif slash >= 2 and lines[x][y] == '실':
            fail += 1  
    name = ''.join(name)
    #기록
    if slash + 1 == fail:
        df.loc[name,date] = 'x'
    elif fail > 0:
        df.loc[name,date] = 'ox'
    elif fail == 0:
         df.loc[name,date] = 'o'
```

### 새로운 날짜에 결측값이 있으면 '?'기록

```python
df = df.fillna('?')
df.head()
```

### 결과 정리

```python
value = df[Today]
o_num = 0
x_num = 0
ox_num = 0
for x in range(len(value)):
    if(value[x]=='o'):
        o_num += 1
    elif(value[x] == 'x'):
        x_num += 1
    elif(value[x]=='ox'):
        ox_num += 1
df.loc['O인원',Today] = o_num
df.loc['OX인원',Today] = ox_num
df.loc['X인원',Today] = x_num
df.loc['?인원',Today] = 23-(o_num + ox_num + x_num)
```

```python
df.to_excel("결과.xlsx")
```
